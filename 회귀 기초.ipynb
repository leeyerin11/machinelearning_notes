# 단순회귀

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/user/Desktop/학교 활동/쿠글/쿠글 여름 스터디/2주차 자료/part7/auto-mpg.csv",header=None)
df
```

```python
df.columns=['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','name']
df.head()
```

```python
df.info()  ##int형 float형 섞여있음, horsepower 왜 object?(결측치가 ?로 표현되어 있음)
```

```python
df.describe()
```

```python
df['horsepower'].unique()
```

```python
df['horsepower'].replace('?',np.nan,inplace=True)
```

```python
df.dropna(subset=['horsepower'],axis=0,inplace=True)
df['horsepower']=df['horsepower'].astype('float')

df.describe()
```

```python
df.plot(kind='scatter',x='weight',y='mpg',figsize=(10,5))
```

```python
fit=plt.figure(figsize=(10,5))
sns.regplot(x='weight',y='mpg',data=df)
plt.show()
```

```python
sns.jointplot(x='weight',y='mpg',kind='reg',data=df)
plt.show()
```

```python
grid_df=sns.pairplot(df)
plt.show()
```

```python
X=df[['weight']]   ##행렬 곱을 위해 2차원으로 써주기
y=df['mpg']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=111)

print(len(X_train))
print(len(y_train))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)

r_square=lr.score(X_test,y_test)
print(r_square)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
y_hat=lr.predict(X)

plt.figure(figsize=(10,5))
ax1=sns.distplot(y,hist=False,label='y')
ax2=sns.distplot(y_hat,hist=False,label='y')

plt.show()
```

# 다항회귀

```python
from sklearn.preprocessing import PolynomialFeatures   #다항식 변환 위한 모듈
```

```python
poly=PolynomialFeatures(degree=2)  #2차항
X_train_poly=poly.fit_transform(X_train)
```

```python
print(X_train.shape)
print(X_train_poly.shape)
```

```python
pr=LinearRegression()
pr.fit(X_train_poly,y_train)
```

```python
X_test_poly=poly.fit_transform(X_test)
r_square=pr.score(X_test_poly,y_test)
print(r_square)
```

```python
y_hat_test=pr.predict(X_test_poly)

fig=plt.figure(figsize=(10,5))
ax=fig.add_subplot(1,1,1)
ax.plot(X_train,y_train,'o',label='Train Data')
ax.plot(X_test,y_hat_test,'r+',label='Predicted Value')
ax.legend(loc='best')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
X_poly=poly.fit_transform(X)
y_hat=pr.predict(X_poly)

plt.figure(figsize=(10,5))
ax1=sns.distplot(y,hist=False,label="y")
ax2=sns.distplot(y_hat,hist=False,label="y_hat",ax=ax1)
plt.show()
```
